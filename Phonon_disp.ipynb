```python
import os
import shutil
import subprocess

# === USER CONFIGURATION ===
num_processors = 6
atoms = ['Ga', 'As']
dim = [2, 2, 1] 
n_displacements = 8 # 2*2*2 *num 2  
input_path = "/global/cfs/projectdirs/m3845/Diego_projects/Siepho_disp_calc/"
pot_path = os.path.join(input_path, "potentials")
fdf_filename = "2D-GaAs.fdf"
pot_Ga = "Ga.psf"
pot_As = "As.psf"
phonopy_exec = "phonopy"
siesta_exec = "/global/homes/d/dajuarez/.conda/envs/siesta_env/bin/siesta"


# === SETUP ENVIRONMENT ===
os.environ["PATH"] = os.path.dirname(siesta_exec) + ":" + os.environ["PATH"]
os.chdir(input_path)

# === STEP 1: GENERATE DISPLACEMENTS ===

phonopy_command = f'{phonopy_exec} --siesta -d --dim="{dim[0]} {dim[1]} {dim[2]}" -c {fdf_filename}'
subprocess.run(phonopy_command, shell=True, check=True)

# # === STEP 2: EXTRACT FDF BLOCKS ===
start_keywords = ["%block ChemicalSpeciesLabel", "%endblock ChemicalSpeciesLabel"]
end_keywords = [
    "%block GeometryConstraints", "%endblock GeometryConstraints",
    "PAO.BasisSize", "PAO.EnergyShift", "XC.Functional", "XC.Authors",
    "MaxSCFIterations", "MeshCutoff", "DM.Tolerance", "DM.MaxingWeight",
    "WriteForces", "MD.UseSaveXV", "SolutionMethod",
    "%block kgrid_Monkhorst_Pack", "%endblock kgrid_Monkhorst_Pack",
    "MD.TypeOfRun", "MD.NumCGsteps", "MD.VariableCell",
    "MD.MaxForceTol", "MD.TargetPressure"
]

def extract_properties(file_path):
    start_props, end_props = [], []
    with open(file_path, 'r') as f:
        lines = f.readlines()
    in_start, in_end = False, False
    for line in lines:
        if any(k in line for k in start_keywords):
            in_start = True
        if in_start:
            start_props.append(line)
            if "%endblock" in line:
                in_start = False
        if any(k in line for k in end_keywords):
            in_end = True
        if in_end:
            end_props.append(line)
    return start_props, end_props

start_props, end_props = extract_properties(fdf_filename)

# # === STEP 3: MODIFY AND RUN SIESTA FOR EACH DISPLACEMENT ===
for i in range(1, n_displacements + 1):
    sc_file = f"supercell-{i:03d}.fdf"  
    disp_dir = f"disp_{i:03d}"           
    os.makedirs(disp_dir, exist_ok=True)

    with open(sc_file, 'r') as f:
        content = f.read()
    new_content = "".join(start_props) + "\n# Original supercell content\n" + content + "\n" + "".join(end_props)
    with open(sc_file, 'w') as f:
        f.write(new_content)

    shutil.move(sc_file, os.path.join(disp_dir, sc_file))
    shutil.copy(os.path.join(pot_path, pot_Ga), os.path.join(disp_dir, pot_Ga))
    shutil.copy(os.path.join(pot_path, pot_As), os.path.join(disp_dir, pot_As))

    cmd = f"mpirun -np {num_processors} {siesta_exec} < {sc_file} | tee output.txt"
    subprocess.run(cmd, shell=True, check=True, cwd=disp_dir)
    print(f"[] SIESTA run complete for {disp_dir}")

# === STEP 4: GENERATE FORCE_SETS ===
fa_paths = " ".join([f"disp_{i:03d}/siesta.FA" for i in range(1, n_displacements + 1)])
forces_command = f"{phonopy_exec} --siesta -f {fa_paths}"
subprocess.run(forces_command, shell=True, check=True)
command1 = 'phonopy --siesta -c 2D-GaAs-GGA-RL-2.fdf -p -s band.conf --dim="2 2 2"'
command2 = 'phonopy --siesta -c 2D-GaAs-GGA-RL-2.fdf -p  band.conf' 
os.chdir('./')
os.system(command1)
os.system(command2)
print("All calculations complete. FORCE_SETS generated.")
```
